### АРМ: автоматизированное рабочее место. My software

Недавно обживал новый ноутбук и устанавливал различные программы. Вот краткий список самого необходимого.
Работаю я под управлением ОС Windows 8.1 x64. Этого, в принципе, мне хватает за глаза. Найти материал, прочитать его, оставить в закладках, скачать данные и сделать расчёты, написать отчёт, сконверитровать pdf, выложить материалы в блог или на сайт.

### Офисные программы + текстовые редакторы

1. [Notepad++](http://notepad-plus-plus.org/download/v6.6.9.html) - отличный текстовый редактор, замена Блокноту

2. MS Office (Word-Excel-PowerPoint) 2010-2013 - де-факто стандарт офисных процессоров

3. [Geany](http://www.geany.org/Download/Releases) - замечательный текстовый редактор, удобный для работы с языками программирования (ставятся отдельно)

4. [Adobe Reader](http://www.adobe.com/ru/products/reader.html) - чтение и печать pdf файлов

5. [WinDjView](http://windjview.sourceforge.net/ru/) - чтение и печать djvu файлов 

6. [TexStudio](http://texstudio.sourceforge.net/) - редактор для TeX


### Навигация в ОС

1. [Total Commander](http://www.ghisler.com/) - двухоконный менеджер

2. [Yandex Browser](http://browser.yandex.ru/#main) - браузер на основе Chromium от Яндекса

3. [Punto Switcher](http://punto.yandex.ru/) - переключение регистра, языка и перевод текста, набранного в другой языковой раскладке

### Социальные программы

1. [Skype](http://www.skype.com/ru/) - общение с коллегами и друзьями зарубежом и по России

2. [Github Windows](https://windows.github.com/) - визуальный интерфейс для git-технологии для портала GitHub
 
### Научные пакеты

1. [Python in Anaconda distribution](https://store.continuum.io/cshop/anaconda/) - бесплатная версия научной сборки python от Continuum Analytics

2. [MinGW](http://sourceforge.net/projects/mingw/files/) - gfortran (и не только) под Windows

3. [Mendeley](http://www.mendeley.com/) - отличный бесплатный менеджер научной библиографии

4. [TexLive](http://www.tug.org/texlive/acquire-netinstall.html) - мультиплатформенная сборка LateX

### Системные утилиты

1. [Daemon Tools Lite](http://www.disk-tools.com/download/daemon) - чтение образов диска

2. [Ccleaner](http://www.piriform.com/ccleaner) - утилита очистки реестра, временных файлов и управление автозагрузками

3. [RevoUnistaller](http://www.revouninstaller.com/revo_uninstaller_free_download.html) - глубокое удаление программ из реестра
